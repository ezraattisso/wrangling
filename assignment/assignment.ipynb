```python
import numpy as np # Import the numpy package into your workspace
import matplotlib.pyplot as plt # Import matplotlib into your workspace
import pandas as pd  # Import the pandas package into your workspace

## Q2.1
sharks_url = "https://raw.githubusercontent.com/ezraattisso/wrangling/refs/heads/main/data/sharks.csv"

sharks_df = pd.read_csv(sharks_url, low_memory=False)

sharks_df.head()

## Q2.2

sharks_df = sharks_df.dropna(axis=1, how="all")

# this function will help delete any rows that are completely empty.

## Q2.3

sharks_df["Year"] = sharks_df["Year"].astype(int)

# this is us fixing the column to show the values as integers, and the below query is tof ilter based on the year 1940.

sharks_df = sharks_df[sharks_df["Year"] >= 1940]


## Q2.4

sharks_df["A"] = sharks_df["Age"].fillna(0)

sharks_df['Age'] = sharks_df['Age'] # Categorical variable example
var = 'Age'

print(sharks_df[var].hist(), '\n')


## Q2.5

sharks_df = sharks_df.rename(columns={"Unnamed: 9": "Sex"})

gender_proportion = sharks_df["Sex"].value_counts(normalize=True)

# this function will determine the percentages of the Ms, Fs, and NAs.

print(gender_proportion)


## Q2.6

sharks_df.loc[sharks_df['Type'].str.contains('Provoked'), 'Type'] = 'Provoked'
sharks_df.loc[sharks_df['Type'].str.contains('Unprovoked'), 'Type'] = 'Unprovoked'
sharks_df.loc[sharks_df['Type'].str.contains('NaN'), 'Type'] = 'Unknown'

## Q2.7

sharks_df.loc[sharks_df['Fatal (Y/N)'].str.contains('N'), 'Fatal (Y/N)'] = 'N'
sharks_df.loc[sharks_df['Fatal (Y/N)'].str.contains('Y'), 'Fatal (Y/N)'] = 'Y'
sharks_df.loc[sharks_df['Fatal (Y/N)'].str.contains('NaN'), 'Type'] = 'Unknown'

# this query finds where the columns contain each characters and replaces it with the desired one we want.


## Q2.8

likely_df = sharks_df[sharks_df["Type"] == "Unprovoked", sharks_df["Sex"]]

likely_df.head()

# this would show us the columns of unprovoked attacks alongside the genders associated, and from there we would be able to see if there was a trend between the two.




## Q2.9

sharks_df["species_words"] = sharks_df["Species"].str.lower().str.split()
# this is you converting text to lowercase and split each entry into a list of words


white_shark_attacks = sharks_df[sharks_df["species_words"].apply(lambda x: "white" in x)]
# this is you counting rows where "white" appears in the word list



proportion_white_sharks = len(white_shark_attacks) / len(sharks_df)
# Calculate proportion here.

print(f"Proportion of attacks by white sharks: {proportion_white_sharks:.2%}")


```
